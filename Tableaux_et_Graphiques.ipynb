## **Tableaux et graphiques**

Dans ce chapitre, nous allons réaliser des graphiques à partir de données stockées dans des tableuax, à l'aide des modules NumPy et Matplotlib.


Commençons par importer les module nécessaires.


```
import numpy as np
import matplotlib.pyplot as plt
```

Créons un tableau à partir d'une liste et traçons son contenu.

```
t = np.array([7,1,-1,5])
plt.plot(t)
```

Créons un nouveau tableau x qui contiendra 100 valeurs d'abscisses régulièrement espacées dans l'intervalle -10 : +10.

Créons ensuite un tableau y qui stockera le sinus de chacune des valeurs de x.

Traçons enfin la fonction sinus en utilisant les valeurs x en abscisse et de y en ordonnée.

```
x = np.linspace(-10,10,100)
y = np.sin(x)
plt.plot(x, y, label='$y = \sin(x)$')
plt.legend()
```

Visualisons la distribution empirique des tirages sous forme d’histogramme (1000 valeurs, réparties sur R barres), en fonction de la méthode de tirage aléatoire

```
a = np.random.standard_normal(10000)
h1 = plt.hist(a, 40)
b = np.random.random_sample(10000)
plt.figure()
h2 = plt.hist(b, 40)
```

Visualisons enfin le contenu d’une matrice de 32 cases par 32 cases remplies de valeurs tirées aléatoirement en suivant une loi normale standard, comme une image.

```
C = np.random.standard_normal((32,32))
plt.imshow(C, interpolation='nearest')
plt.colorbar()
```
