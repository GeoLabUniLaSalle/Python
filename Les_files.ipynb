## **Les files**

Dans ce chapitre, nous allons étudier le fonctionnement des files qui sont des structures beaucoup utilisées en informatique.

Les files, ou **queue**, fonctionnent comme une file d'attente. Lorsqu'une personne rejoint une file d'attente, elle se positionne en dernière position, et c'est la personne qui a rejoint la file d'attente en premier qui sera servi d'abord. On appelle cela la structure **FIFO** (First In, First Out).

Nous allons voir ici comment coder le fonctionnement d'une file en Python. La file permettra : d'accueillir un nouvel élément, de supprimer un élément, de connaitre l'élément positionné au dessus de la file, et le nombre d'éléments appartenant à la file.

```
class File:
  
  def __init__(self):
    self.elements = []
    
  def enfile(self, element):
    self.elements.append(element)

  def defile(self):
    if self.elements:
      return self.elements.pop(0)
  
  def taille(self):
    return len(self.elements)

  def sommet(self):
    if len(self.elements)==0:
         raise Exception("La file est vide")
    return self.elements[0]
    
  def __str__(self):
    out = str(self.elements[0])
    for x in self.elements[1:]:
      out = str(x) + " -> " + out
    out = "File : " + out
    return out

queue = File()
queue.enfile(4)
queue.enfile(9)
queue.enfile(1)

print(queue)

queue.defile()

print('Sommet',queue.sommet())
print('Taille',queue.taille())

queue.enfile(3)

print(queue)
```

Exercice d'application : **Le jeu de Josèphe**.

Soit une liste représentant un ensemble de personnes, incluant Josèphe, disposées en cercle. Tour à tour, chaque personne sera éliminée, en comptant de 3 en 3 à partir de la première personne.

Capture d’écran 2021-03-02 à 23.40.05.png

Le problème est de savoir où Josèphe doit se placer dans le cercle pour être éliminé en dernier. Nous allons utiliser une file pour résoudre ce problème.

```
cercle = ['A','B','C','D','E','F','G']

def dernier_elimine(c):
  f = File()
  for i in c:
    f.enfile(i)
  while(f.taille()!=0):
    f.enfile(f.defile())
    f.enfile(f.defile())
    sortie = f.defile()
  return sortie

print(dernier_elimine(cercle))
```
