# **Les piles et files**

Dans ce chapitre, nous allons étudier le fonctionnement des piles et des files qui sont des structures beaucoup utilisées en informatique.

## **Les piles**

Les piles, ou **stack**, fonctionnent commune pile d'assiette. Lorsqu'on ajoute une assiette à la pile, on la place au sommet, et lorsque l'on sort une assiette de la pile, on prend celle qui est au sommet. On appele cela la structure **LIFO** (Last In, First Out).

Nous allons voir ici comment coder le fonctionnement d'une pile en Python. La pile permettra : d'accueillir un nouvel élément, de supprimer un élément, de connaitre l'élément positionné au dessus de la pile, et le nombre d'éléments appartenant à la pile.

```
class Pile:
  
  def __init__(self):
    self.elements = []
    
  def empile(self, element):
    self.elements.append(element)

  def depile(self):
    if self.elements:
      return self.elements.pop()
  
  def taille(self):
    return len(self.elements)

  def sommet(self):
    if len(self.elements)==0:
         raise Exception("La pile est vide")
    return self.elements[-1]
    
  def __str__(self):
    out = str(self.elements[0])
    for x in self.elements[1:]:
      out = str(x) + " -> " + out
    out = "Pile : " + out
    return out

stack = Pile()
stack.empile(4)
stack.empile(9)
stack.empile(1)

print(stack)

stack.depile()

print('Sommet',stack.sommet())
print('Taille',stack.taille())

stack.empile(3)

print(stack)
```

Exercice d'application : **La gare de triage**.

Soit une liste représentant un train dont les élements sont des wagons, qui sont chacun associé à un nombre.

Nous souhaitons réorganiser le train pour que les wagons associés à un nombre impair soient positionnés avant les wagons représentés par un nombre pair.

Capture d’écran 2021-03-02 à 22.51.56.png

Pour réaliser cette opération, nous disposons d'une gare d'entrée et d'une zone d'attente, et chaque wagon provenant de la gare d'entrée peut être envoyé directement en gare de sortie ou être stocké momentanément dans la zone d'attente.

Capture d’écran 2021-03-02 à 22.53.24.png
Capture d’écran 2021-03-02 à 23.12.55.png

Nous allons utiliser une pile pour résoudre ce problème.



```
train = [7,16,4,3,9,6,11]

def tri_wagons(t):
  sortie = []
  attente = Pile()
  for i in t:
    if i%2==0:
      attente.empile(i)
    else:
      sortie.append(i)
  while(attente.taille()!=0):
    sortie.append(attente.depile())
  return sortie

print(tri_wagons(train))
```

## **Les files**

Les piles, ou **queue**, fonctionnent commune file d'attente. Lorsqu'un personne rejoint une file d'attente, elle se positionne en dernière position, et c'est la personne qui a rejoint la file d'attente en premier qui sera servi d'avord. On appele cela la structure **FIFO** (Fast In, First Out).

Nous allons voir ici comment coder le fonctionnement d'une file en Python. La file permettra : d'accueillir un nouvel élément, de supprimer un élément, de connaitre l'élément positionné au dessus de la file, et le nombre d'éléments appartenant à la file.

```
class File:
  
  def __init__(self):
    self.elements = []
    
  def enfile(self, element):
    self.elements.append(element)

  def defile(self):
    if self.elements:
      return self.elements.pop(0)
  
  def taille(self):
    return len(self.elements)

  def sommet(self):
    if len(self.elements)==0:
         raise Exception("La file est vide")
    return self.elements[0]
    
  def __str__(self):
    out = str(self.elements[0])
    for x in self.elements[1:]:
      out = str(x) + " -> " + out
    out = "Pile : " + out
    return out

queue = File()
queue.enfile(4)
queue.enfile(9)
queue.enfile(1)

print(queue)

queue.defile()

print('Sommet',queue.sommet())
print('Taille',queue.taille())

queue.enfile(3)

print(queue)
```

Exercice d'application : **Le jeu de Josèphe**.

Soit une liste représentant un ensemble de personnes, incluant Josèphe, disposées en cercle. Tour à tour, chaque personne sera éliminée, en comptant de 3 en 3 à partir de la première personne.

Capture d’écran 2021-03-02 à 23.40.05.png

Le problème est de savoir où Josèphe doit se placer dans le cercle pour être éliminé en dernier. Nous allons utiliser une file pour résoudre ce problème.

```
cercle = ['A','B','C','D','E','F','G']

def dernier_elimine(c):
  f = File()
  for i in c:
    f.enfile(i)
  while(f.taille()!=0):
    f.enfile(f.defile())
    f.enfile(f.defile())
    sortie = f.defile()
  return sortie

print(dernier_elimine(cercle))
```
