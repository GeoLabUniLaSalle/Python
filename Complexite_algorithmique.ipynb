# **Complexité algorithmique**

Dans ce chapitre, nous allons étudier la **complexité algorithmique** de plusieurs problèmes.

Notre objectif est de savoir évaluer si les solutions que nous proposons pour résoudre un problème peuvent être trouvées dans un temps raisonnable.

## **Complexité en temps (O)**

Lorsque l'on cherche à évaluer la complexité alrogithmique d'un problème, il faut analyser les ressources nécessaires pour le résoudre. Ces ressources dépendent de la taille des données fournies en entrée, et la complexité est généralement calculée en fonction de *n*, où *n* est la taille des données en entrée.

Pour analyser la complexité algorithmique, nous pouvons prendre en compte la complexité en temps (d'exécution), et en espace (mémoire utilisé). Ici, nous nous concentrons sur la complexité en temps d'exécution, qui est estiméz en comptant le nombre d'opérations élémentaires (affectations, instructions arithmétiques, logiques, évaluation de conditions, etc.) réalisées par l'algorithme étudié.

Les opérations élémentaires sont exécutées par le processeur (CPU ou GPU) de l'ordinateur, et sa fréquence nous indique le temps de calcul. Par exemple, un processeur cadencé à 3 GHz effectue 3 milliards d'opérations par seconde. Le temps d'exécution dépend donc de *n*, et de l'ordinateur sur lequel le problème est traité.

Lorsque l'on analyse la complexité en temps d'un algorithme, nous pouvons étudier 3 cas : le meilleur cas, le cas moyen, et le pire des cas.

Prenons un exemple. Nous cherchons si une valeur est présente dans une liste de taille *n* correspondant au nombre d'éléments qu'elle contient, en utilisant la fonction de recherche suivante :

```python
nombres = [5,8,1,4,2] # n = 5

def recherche(x,l):
  """
  Fonction booléenne qui teste si x est un élément de la liste l
  """
  for i in l:
    if i==x:
      return True
  return False
```

Le meilleur cas est celui où le nombre recherché est le premier de la liste (5).

```python
nombres = [5,8,1,4,2]

nb = 5
if recherche(nb,nombres):
  print('Le nombre',nb,'est présent')
else:
  print('Le nombre',nb,"n'est pas présent")
```

Ici, il a suffi d'une itération pour résoudre le problème

Le cas moyen est le nombre d'itérations moyen nécessaires pour résoudre le problème (si on imagine qu'on exécute la recherche un grand nombre de fois avec des valeurs tirées aléatoirement et équitablement distribuées). Nous pouvons rapprocher ce cas moyen à la recherche du nombre présent au milieu de la liste (1).

```python
nombres = [5,8,1,4,2]

nb = 1
if recherche(nb,nombres):
  print('Le nombre',nb,'est présent')
else:
  print('Le nombre',nb,"n'est pas présent")
```

Nous avons besoin de 3 itérations pour résoudre le problème. C'est ce qui se produira généralement lorsque nous utiliserons cette fonction : le nombre moyen d'itérations sera égal à la moitié de la taille de la liste (*n*/2).

Le pire des cas est celui où la valeur recherchée est située en dernière position (2).

```python
nombres = [5,8,1,4,2]

nb = 2
if recherche(nb,nombres):
  print('Le nombre',nb,'est présent')
else:
  print('Le nombre',nb,"n'est pas présent")
```

Il nous faut ici 5 itérations, soit le nombre d'éléments de la liste (*n*), pour résoudre le problème.

Lorsque l'on parle de complexité algorithmique, on parle du pire des cas, et on note cette complexité "Grand O". Ici, on dit que la complexité de cet algorithme de recherche est "Grand O de n", ou "de l'ordre de n", puisque dans le pire des cas il faut *n* opérations pour résoudre le problème, et on note la complexité de cet algorithme ainsi : *O(n)*.

## **Exemples d'algorithmes**

## Déterminer si un nombre est pair.

```python
def pair(x):
  """
  Fonction booléenne qui teste si x est pair ou non
  """
  return x%2==0

print(pair(7))
print(pair(8))
```

Ici, le nombre d'opérations réalisé par la fonction *pair* est de 1. On dit que cette fonction s'execute en **temps constant**. Sa compléxité est *O(1)*.

## Recherche Dichotomique.

La recherche dichotomique consiste à dire si un nombre est présent ou non dans une liste **triée**. La recherche dichotomique consiste à regarder si l'élément recherché est la valeur du milieu de la liste. Si ce n'est pas le cas, on poursuit la recherche dans la moitié pertinent de la liste. On répète l'opération jusqu'à ce que le nombre soit trouvé, ou qu'il ne reste plus d'éléments à examiner.

```python
def recherche_dichotomique(x,l):
  """
  Fonction booléenne qui teste si x est présent dans l, une liste triée en ordre croissant
  """
  if l==[]:
    return False
  milieu = len(l)//2
  if x==l[milieu]:
    return True
  elif x>l[milieu]:
    return recherche_dichotomique(x,l[milieu+1:])
  else:
    return recherche_dichotomique(x,l[:milieu])

print(recherche_dichotomique(11,[1,4,5,5,7,8,11,14,14,19]))
```

La recherche dichotomique s'écrit de manière récursive.

On cherche à déterminer *k*, le nombre total d'opération.

*n*, le nombre d'éléments de la liste, est divisé par 2 à chaque appel de la fonction *recherche_dichotomique*. La taille finale de la liste est donc de n/2ᵏ. Etant donné que la plus petite taille finale de la liste est de 1, on a 1 ≤ n/2ᵏ. En multipliant par 2ᵏ, on obtient 2ᵏ ≤ n. En appliquant un log décimal, on obtient k.log(2) ≤ log(n). Enfin, en divisant par log(2), nous avons k ≤ log(n)/log(2) = log₂(n).

On dit que cette fonction s'execute en **temps logarithmique**. Sa compléxité est *O(log₂(n))*.

## Tri par insertion.

Le tri par insertion consiste à trier les éléments d'une liste. Il n'est efficace que sur des entrées de petite taille. Nous utilisons naturellement cette technique de tri quand nous jouons aux cartes. Lorsque nous recevons des cartes une par une, nous les trions au fur et à mesure en les intercalant à leur bonne position parmi les cartes déjà piochées.

```python
def tri_insertion(l):
  """
  Fonction qui trie la liste l en ordre croissant
  """
  for i in range(len(l)):
    x = l[i]
    j = i
    while j > 0 and l[j-1] > x :
      l[j] = l[j-1]
      j-=1
    l[j] = x
  return l

print(tri_insertion([1,4,5,5,7,8,11,14,14,19]))
```

Dans le pire des cas, les valeurs en entrée sont triées en ordre décroissant. Prenons un exemple pour calculer la complexité algorithmique de cette fonction, avec la liste suivante :

```python
l = [5,4,3,2,1]

def tri_insertion(l):
  """
  Fonction qui trie et affiche la liste l en ordre croissant
  """
  for i in range(len(l)):
    x = l[i]
    j = i
    while j > 0 and l[j-1] > x :
      print('On echange la valeur',x,'avec la valeur',l[j-1])
      l[j] = l[j-1]
      j-=1
    l[j] = x
    print(l)

tri_insertion(l)
```

Dans le pire des cas : Quand on traite le premier élément, nous avons 0 décalage à réaliser. Pour le second élément, nous avons 1 décalage à réaliser. Pour le troisième élément, nous avons 2 décalages à réaliser, etc.

Le nombre total de décalages potentiels est de 0 + 1 + 2 + 3 + 4. En généralisant à une liste de taille *n*, nous avons S opérations à réaliser :

S = 1 + 2 + 3 + ... + n-3 + n-2 + n-1 

Ecrivons cette somme de manière différente :

S' = n-1 + n-2 + n-3 + ... + 3 + 2 + 1

Si on calcule S + S', on obtient :

S + S' = n + n + n + ... + n + n + n

2S = n(n-1)

S = n(n-1)/2

S = n²/2 + n/2

On dit que cette fonction s'execute en **temps quadratique**. Sa compléxité est *O(n²)*.

## Problèmes d'optimisation

Prenons un problème d'optimisation qui consiste à programme une caisse automatique pour qu'elle rende la monnaie au client de façon optimale, c'est à dire en utilisant le moins de pièces et billets possible.

La valeur des pièces et billets dont nous disposons est contenue dans une liste :

```python
l = [1,2,5,10,20,50,100]
```

Nous disposons d'une infinité de chaque valeur.

Quelle est la meilleure façon de rendre la monnaie à un client qui réalise un achat de 38 euros avec un billet de 100 euros ?



1.   **Force brute**

La méthode par force brute consiste à générer toutes les combinaisons possibles de rendu de monnaie, puis de déterminer parmi ces combinaisons laquelle nécessite le moins de pièces et billets.

La méthode par force brute permet de trouver la solution optimale, mais a une complexité en temps qui est exponentielle. Elle ne peut être envisagée que si la taille de l'entrée est petite.

2. **L'algorithme Glouton**

La stratégie gloutonne consiste à trouver la solution optimale locale à chaque étape de la résolution, en espérant ainsi trouver la solution optimale globale.

Si on applique une stratégie gloutonne à notre problème de rendu de monnaie, nous allons commencer par rendre la pièce ou le billet ayant la plus grand valeur possible (inférieure ou égale à la somme à rendre). On répète cette opération jusqu'à ce que la somme à rendre soit nulle.

Voici une implémentation de cet algorithme :

```python
l = [1,2,5,10,20,50,100]
r = 65

def rendu_glouton(l,r):
  """
  Retourne la solution gloutonne au problème du rendu de monnaie
  """
  m=[]
  while r>0:
    for i in l[::-1]:
      if r-i>=0:
        m.append(i)
        r-=i
        break
  return m

m = rendu_glouton(l,r)
print("Le plus petit nombre de pieces et billets à rendre pour obtenir",r,"euros est de",len(m),"en utilisant la combinaison",m)
```

Attention : Cette solution n'est pas forcément optimale.

Si nous ne disposons plus de billets de 5 et 10 euros et que l'on doit rendre 65 euros, on obtient :

```python
l = [1,2,20,50,100]
r = 65
m = rendu_glouton(l,r)
print("Le plus petit nombre de pieces et billets à rendre pour obtenir",r,"euros est de",len(m),"en utilisant la combinaison",m)
```

Alors que la solution optimale consiste à rendre 3 billets de 20 euros et un billet de 5 euros, soit 4 éléments comparativement aux 9 obtenus avec notre algorithme glouton.

L'algorithme glouton nous donne une solution optimale locale qui n'est pas forcément la solution optimale globale.

Dans certains cas, cet algorithme ne trouve pas du tout de résultat. Si nous ne disposons plus de pièces de 1 euro et que l'on doit rendre 8 euros, on obtient une boucle infinie :

```python
l = [2,5,10,20,50,100]
r = 8
m = rendu_glouton(l,r)
print("Le plus petit nombre de pieces et billets à rendre pour obtenir",r,"euros est de",len(m),"en utilisant la combinaison",m)
```

On commence par rendre un billet de 5 euros, puis une pièce de 2 euros, puis aucune valeur ne nous permet d'atteindre la somme de 8 euros.

La solution ici était d'utiliser 4 pièces de 2 euros.

**3. Comment obtenir une solution optimale à un problème ?**

Cela dépend du problème.

Les algorithmes en **force brute** sont les plus efficaces sur de petites entrées.

Les **algorithmes gloutons** peuvent produire d'excellents résultats dans certains cas ou ne pas convenir du tout. Ils sont efficaces quand les poids des objets traités sont proches, mais moins efficaces dans le cas contraire parce qu'ils n'ont pas une vision globale du problème.

D'autres solutions comme la **programmation dynamique** peuvent être efficaces. La programmation dynamique consiste à décomposer le problème en sous-problèmes à traiter du plus petit au plus grand en stockant les résultats intermédiaires.
