## **Les séquences : slicing**

Dans ce chapitre, nous allons prendre en main les outils de slicing qui s'utilisent sur les **séquences**.

Prenons une variable de type str, les str étant des séquences.

```
s = 'hello, world !'
```

Voici une représentation du contenu de s :
Capture d’écran 2021-01-13 à 15.25.11.png

Il est possible extraire une sous-séquence à partir d'une séquence à partir d'un indice de début (a), et d'un indice de fin (b) :

Slicing de s :

Capture d’écran 2021-01-13 à 16.05.01.png

```
print(s[0:5]) # De l'objet d'indice 0 (inclu) à l'objet d'indice 5 (exclu)
print(s[7:12]) # De l'objet d'indice 7 (inclu) à l'objet d'indice 12 (exclu)
print(s[:5]) # Du début jusqu'à l'objet d'indice 5 (exclu)
print(s[7:]) # De l'objet d'indice 7 (inclu) jusqu'à la fin
print(s[:]) # Du début à la fin
```

Il est également possible d'ajouter un pas (c) :

Capture d’écran 2021-01-13 à 16.04.38.png

```
print(s[0:10:2])
print(s[::2])
print(s[:10:3])
print(s[4::3])
```

Attention : il n'est pas possible de cibler un objet de la séquence dont l'indice est en dehors de la séquence :

```
s[100]
```

Il est par contre possible de slicer avec des bornes dont les valeurs sortent de la séquence :

```
print(s[4:100]) # Retourne la sous-séquence allant de l'objet d'indice 4 (inclu) jusqu'à la fin de la séquence
print(s[100:1000]) # Ne retourne rien parce qu'il n'y a pas de valeurs entre les indices 100 et 1000
```

Les séquences sont naturellement pourvues d'indices négatifs :

Capture d’écran 2021-01-13 à 16.16.19.png

Il est possible d'accéder aux éléments de s par leurs indices négatifs :

```
print(s[-13])
print(s[-1]) # Technique usuelle pour atteindre le dernier élément d'une séquence
```

Les outils de slicing vus précédemment fonctionnent de la même manière avec des valeurs négatives, qui peuvent être combinées avec des valeurs positives :

```
print(s[-10:-4]) # Sous-séquence comprenant les objets d'indice -10 (inclu) à -4 (exclu)
print(s[:-6]) # Sous-séquence comprenant les objets allant du début à celjui d'indice -6 (exclu)
print(s[::-1]) # Tous les éléments, avec un pas de -1, donc de droite à gauche
print(s[8:2:-1]) # Sous-séquence comprenant les objets d'indice 8 (inclu) à 2 (exclu), de droite à gauche
print(s[8::-1]) # Sous-séquence comprenant les objets d'indice 8 (inclu) jusqu'à la fin, de droite à gauche
```
