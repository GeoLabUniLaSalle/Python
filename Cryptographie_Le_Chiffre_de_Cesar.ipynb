# **Introduction à la cryptographie**

Dans ce chapitre, nous allons introduire la notion de **cryptograhie** et découvrir quelques algorithme de **chiffrement faible**

Le chiffrement est un procédé de cryptographie qui consiste à rendre un document illisible à toute personne ne possédant pas la **clé** de déchiffrement.

Ici, nous allons découvrir le fonctionnement de l'**algorithme** de **César**. Cet algorithme, utilisé par Jules César pour ses correspondances privées, emploie un chiffrement par substition, comme **ROT13**. Ici cependant, le décalage n'est pas forcément de *13*, mais de *x*, ou *x* est la *clé de codage*.

Réalisons maintenant la fonction permettant de code un texte avec une clé au choix (à entrer en paramètre) :

```python
def Cesar(x, cle):
  """
  Retourne x chiffré avec le code de César et le chiffre cle
  """
  lower = ascii_lowercase
  upper = ascii_uppercase
  code=""
  for i in x:
    if i in lower:
      code += lower[(lower.index(i)+cle)%26]
    elif i in upper:
      code += upper[(upper.index(i)+cle)%26]
    else:
      code += i
  return code
```

Codons un texte avec cette fonction et la valeur 3 en clé :

```python
code = Cesar('Bonjour !',3)
print(code)
```

Pour décoder, on utilisera la clé -3 (décalage vers la gauche) :

```python
decode = Cesar(code,-3)
print(decode)
```

Cela fonctionne bien entendu avec n'importe quelle clé :

```python
code = Cesar('Bonjour !',21)
print(code)
```

```python
decode = Cesar(code,-21)
print(decode)
```

L'inconvénient de cette technique est que la clé peut facilement être trouvée par force brute. Il suffit de tester les 26 valeurs possibles de décalages pour décoder un texte :

```python
for i in range(26):
  print(Cesar('Erqmrxu !',i))
```
