Voici un premier graphe :

```
G = nx.Graph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
G.add_edge("A","B")
G.add_edge("B","F")
G.add_edge("C","D")
G.add_edge("C","E")
G.add_edge("C","F")
G.add_edge("D","E")
G.add_edge("F","G")
G.add_edge("F","A")
G.add_edge("G","D")

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos)

plt.show()
```

Nous avons créé le graphe G composé de 7 sommets (A, B, C, D, E, F et G) et de 9 arêtes reliant A à B, B à F, etc.

L'instruction *pos = nx.spring_layout(G)* permet de générer un positionnement possible des sommets dans l'espace. En exécutant à nouveau ce bloc d'instruction, le graphe pourra être représenté avec des sommets positionnés autrement. Si l'on supprime cette ligne et que l'on exécute à nouveau ce bloc, le positionnement des sommets sera alors inchangé.

Enfin, nous traçons les sommets avec leur label, et les arêtes.

## **Graphes orientés**

Networks permet de représenter des graphes orientés grâce à la classe Digraph.

En construisant un DiGraph, nous obtenons un graphe orienté. Ici, l'arête reliant A à B est une flèche qui va de A à B. Cela vient de la création de cette arête qui décide de l'orientation : *G.add_edge("A","B")* est différent de *G.add_edge("B","A")*.


```
G = nx.DiGraph() # Nous avons remplacé Graph() par DiGraph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
G.add_edge("A","B")
G.add_edge("B","F")
G.add_edge("C","D")
G.add_edge("C","E")
G.add_edge("C","F")
G.add_edge("D","E")
G.add_edge("F","G")
G.add_edge("F","A")
G.add_edge("G","D")

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos)

plt.show()
```

## **Graphes pondérés**

Il est enfin possible de représenter un graphe pondéré et de faire apparaître le poids de chaque arête. Cela est possible sur des graphes orientés comme non orientés.

```
G = nx.Graph() 
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
G.add_edge("A","B", weight = 12.3)
G.add_edge("B","F", weight = 6.1)
G.add_edge("C","D", weight = 4.2)
G.add_edge("C","E", weight = 17.0)
G.add_edge("C","F", weight = 9.8)
G.add_edge("D","E", weight = 3.3)
G.add_edge("F","G", weight = 6.2)
G.add_edge("F","A", weight = 14.5)
G.add_edge("G","D", weight = 10.6)

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos)

labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)

plt.show()
```

## **Opérations courantes**

Il est possible d'afficher le degré (nombre de voisins) et la liste des voisins d'un sommet :

```
G = nx.Graph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
G.add_edge("A","B")
G.add_edge("B","F")
G.add_edge("C","D")
G.add_edge("C","E")
G.add_edge("C","F")
G.add_edge("D","E")
G.add_edge("F","G")
G.add_edge("F","A")
G.add_edge("G","D")

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos)

plt.show()

print("Degré du sommet D :",G.degree("D"))
print("Liste des sommets voisins du sommet D :",list(G.adj["D"]))
```

Nous pouvons accéder facilement au contenu du graphe :

```
G = nx.Graph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
G.add_edge("A","B", weight = 12.3)
G.add_edge("B","F", weight = 6.1)
G.add_edge("C","D", weight = 4.2)
G.add_edge("C","E", weight = 17.0)
G.add_edge("C","F", weight = 9.8)
G.add_edge("D","E", weight = 3.3)
G.add_edge("F","G", weight = 6.2)
G.add_edge("F","A", weight = 14.5)
G.add_edge("G","D", weight = 10.6)

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos)

labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)

plt.show()

for sommet1, voisins in G.adj.items():
  for sommet2, donnee in voisins.items():
     print(sommet1, sommet2, donnee['weight'])
```

## **Graphes et dictionnaires**

Pour manipuler les graphes en Python, nous les stockons sous la forme de disctionnaires. Un graphe être codé ainsi :

```
D = dict()
D['A'] = {'B','F'}
D['B'] = {'A','F'}
D['C'] = {'D','E','F'}
D['D'] = {'C','E','G'}
D['E'] = {'C','D'}
D['F'] = {'A','B','C','G'}
D['G'] = {'D','F'}
print(D,type(D))
```

Une représentation visuelle de ce graphe est celle d'une matrice :

Capture d’écran 2021-01-28 à 11.24.04.png

Cette matrice peut être lue en lignes ou en colonnes (elle est symétrique), et les 1 indiquent la présence d'une arête entre deux sommets.

Il est possible de représenter ce dictionnaire sous la forme d'un graphe par l'opération suivante :

```
G = nx.Graph(D)

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos)

plt.show()
```

Cette méthode nous permet de nous affranchir du module NetworkX pour traiter un graphe, si on ne souhaite pas l'afficher.

Il est possible de réaliser un graphe orienté de cette manière aussi :

```
D = dict()
D['A'] = {'B'}
D['B'] = {'F'}
D['C'] = {'D','E','F'}
D['D'] = {'E'}
D['E'] = {}
D['F'] = {'A','G'}
D['G'] = {'D'}

import networkx as nx

G = nx.DiGraph(D)

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos)

plt.show()
```

Enfin, il est possible de représenter un graphe pondéré sous la forme de dictionnaire :

```
D = dict()
D['A'] = {'B':{'weight':12.3}}
D['B'] = {'F': {'weight': 6.1}}
D['C'] = {'D': {'weight': 4.2}, 'E': {'weight': 17.0}, 'F': {'weight': 9.8}}
D['D'] = {'E': {'weight': 3.3}}
D['E'] = {}
D['F'] = {'G': {'weight': 6.2}, 'A': {'weight': 14.5}}
D['G'] = {'D': {'weight': 10.6}}

import networkx as nx

G = nx.Graph(D)

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos)

labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)

plt.show()
```

## **Exemple d'utilisation**

Nous souhaitons représenter, sous forme de graphe, un réseau d'émetteurs et de récepteurs. Les émetteurs et récepteurs sont les sommets de notre graphe, et un arête entre deux sommets indiquent que l'émetteur et le récepteur peuvent communiquer.

Voici le réseau que l'on souhaite représenter, où les sommets en rouge sont les émetteurs, et les sommets en jaune sont les récepteurs :

Capture d’écran 2021-03-07 à 12.42.35.png

Voici une  façon de stocker ces informations dans un graphe orienté : Chaque sommet peut être un émetteur ou un récepteur et l'orientation des arêtes nous donne l'information de qui est un émetteur et qui est un capteur. Ici, les arêtes vont des sommets émetteurs vers les sommets récepteurs.

```
D = dict()
D['A'] = {'D','E','G'}
D['B'] = {}
D['C'] = {'B','D','E'}
D['D'] = {}
D['E'] = {}
D['F'] = {'B','E'}
D['G'] = {}

import networkx as nx

G = nx.DiGraph(D)

nx.draw_circular(G,with_labels=True) # Positionne les sommets de manière circulaire

plt.show()
```

Il est possible de visualiser quel sommet est un émetteur et quel sommet est un récepeur ainsi :

```
D = dict()
D['A'] = {'D','E','G'}
D['B'] = {}
D['C'] = {'B','D','E'}
D['D'] = {}
D['E'] = {}
D['F'] = {'B','E'}
D['G'] = {}

emetteur = dict()

# Si le graphe est orienté, on peut écrire
for i in D:
  emetteur[i] = D[i]=={}

'''
Sinon, il est possible de choisir manuellement quel sommet est émétteur et quel sommet est récepteur
emetteur['A'] = True
emetteur['B'] = False
emetteur['C'] = True
emetteur['D'] = False
emetteur['E'] = False
emetteur['F'] = True
emetteur['G'] = False
'''

import networkx as nx

G = nx.DiGraph(D)
color_map=[]
for i in G:
  if(emetteur[i]):
    color_map.append('red')
  else:
    color_map.append('orange')
nx.draw_circular(G,node_color=color_map,with_labels=True) # Positionne les sommets de manière circulaire

plt.show()
```
