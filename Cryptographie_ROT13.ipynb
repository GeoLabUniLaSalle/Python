# **Introduction à la cryptographie**

Dans ce chapitre, nous allons introduire la notion de **cryptograhie** et découvrir quelques algorithme de **chiffrement faible**

Le chiffrement est un procédé de cryptographie qui consiste à rendre un document illisible à toute personne ne possédant pas la **clé** de déchiffrement.

Ici, nous allons découvrir le fonctionnement de l'**algorithme** **ROT13** (*rotate by 13 places*). Cet algorithme consiste à décaler de 13 caractères chaque lettre du texte à chiffrer, et si l'on atteind la dernière lettre ('z'), on repdrend au début ('a').

ROT13.png

Cette procédure sert à **coder** et à **décoder** un texte. En effet, puisqu'il y a 26 lettres dans l'alphabet, il suffit d'appliquer le codage ROT13 sur un texte codé pour le décoder.

Voyons maintenant comment décaler une lettre de 13 caractères.

Observons la position d'une lettre dans l'alphabet :

Ajoutons 13 à cet indice.

```python
lower = ascii_lowercase
lettre = 'j'
indice = lower.index(lettre)
print(lower[indice + 13])
```

```python
lower = ascii_lowercase
lettre = 'x'
indice = lower.index(lettre)
print(lower[indice + 13])
```

Nous observons que si l'indice dépasse 25 (caractère 'z'), nous obtenons un message d'erreur. Nous utilisons le modulo pour éviter ce problème :

```python
lower = ascii_lowercase
lettre = 'x'
indice = lower.index(lettre)
print(lower[(indice + 13)%26])
```

Réalisons maintenant la fonction *ROT13* qui fonctionn avec n'importe quel *str* :

```python
def ROT13(x):
  """
  Retourne x chiffré avec ROT13
  """
  lower = ascii_lowercase
  upper = ascii_uppercase
  code=""
  for i in x:
    if i in lower:
      code += lower[(lower.index(i)+13)%26]
    elif i in upper:
      code += upper[(upper.index(i)+13)%26]
    else:
      code += i
  return code
```

Nous allons maintenant coder le texte 'Bonjour !' :

```python
code = ROT13('Bonjour !')
print(code)
```

Pour décoder ce texte, il suffit de lui appliquer à nouveau la fonction *ROT13* :

```python
decode = ROT13(code)
print(decode)
```
