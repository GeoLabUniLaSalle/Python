# **Introduction aux graphes**

Dans ce chapitre, nous allons introduire la notion de graphe et découvrir le vocabulaire associé

Afin de représenter nos graphes en Python, nous utiliserons le module networkx que nous détaillerons petit à petit.

Un graphe est un ensemble de **sommets** et d'**arêtes** qui relient des sommets entre-eux. Les graphes servent à représenter des relations entre des éléments : villes reliées par des routes, personnes reliées sur un réseau social, etc.

Voici un exemple de graphe :

```
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_node("A") # Création d'un sommet
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
G.add_edge("A","B") # Création d'une arête qui relie les sommets A et B
G.add_edge("B","F")
G.add_edge("C","D")
G.add_edge("C","E")
G.add_edge("C","F")
G.add_edge("D","E")
G.add_edge("F","G")
G.add_edge("F","A")
G.add_edge("G","D")

pos = nx.spring_layout(G) # Génération d'un positionnement des sommets dans l'espace

nx.draw_networkx_nodes(G, pos) # Affichage des sommets
nx.draw_networkx_edges(G, pos, arrows = False) # Affichage des arêtes

plt.show() # Affichage du graphe
```

Ici, nous avons 7 **sommets** représentés par des ronds bleus, et 9 **arêtes** représentées par des lignes noires.

*Remarque : Ce graphe peut être représenté de plusieurs façons, en positionnant les sommets à des endroits différents par exemple, ou avec des arêtes qui se croisent. Cela n'a aucune importance. En exécutant à nouveau le code ci-dessus, vous obtiendrez une autre représentation du même graphe.*

Il est possible de donner des noms aux sommets, et de les afficher :

```
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
G.add_edge("A","B")
G.add_edge("B","F")
G.add_edge("C","D")
G.add_edge("C","E")
G.add_edge("C","F")
G.add_edge("D","E")
G.add_edge("F","G")
G.add_edge("F","A")
G.add_edge("G","D")

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos) # Affichage du nom des sommets
nx.draw_networkx_edges(G, pos, arrows = False)

plt.show()
```

Un **chemin** est une suite d'arêtes qui relie deux sommets.

Plusieurs chemins peuvent relier deux arêtes. Ici, E et F sont par exemple reliés par un chemin passant par C et par un autre chemin passant par D et G.

On dit que le chemin passant par C est de longueur 2 et que celui passant par D et G est de longueur 3. La **longueur** est le nombre d'arrêtes parcourues pour relier deux sommets.

On dit qu'un graphe est **connexe** s'il est d'un seul tenant, c'est à dire que tous les sommets du graphe sont reliés par un **chemin**. Le graphe ci-dessus est donc connexe.

Cela signifie aussi qu'un graphe peut être non connexe :


```
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
G.add_edge("A","B")
G.add_edge("B","F")
G.add_edge("C","D")
G.add_edge("C","E")
G.add_edge("D","E")
G.add_edge("F","A")
G.add_edge("G","D")

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, arrows = False)

plt.show()
```

Un graphe peut même ne pas contenir d'arêtes :

```
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, arrows = False)

plt.show()
```

A l'inverse, si tous les sommets d'un graphe sont reliés par un arête, on dit que le graphe est **complet** :

```
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_edge("A","B")
G.add_edge("A","C")
G.add_edge("A","D")
G.add_edge("B","C")
G.add_edge("B","D")
G.add_edge("C","D")

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, arrows = False)

plt.show()
```

On dit que deux sommets sont **voisins** s'ils sont reliés par une arête.

```
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_edge("A","B")

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, arrows = False)

plt.show()
```

Les sommets A et B sont **voisins** alors que A et C ne sont pas **voisins**.

Le **degré** d'un sommet est égal au nombre de ses voisins. 


```
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_edge("A","B")
G.add_edge("A","C")
G.add_edge("A","D")

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, arrows = False)

plt.show()
```

On dit que A est de **degré** 3 parce qu'il a 3 voisins (B, C, et D), et on note *deg(A) = 3*.

Les sommets B, C, et D sont tous les trois de degré A : *deg(B)=deg(C)=deg(D)=1*.

*Remarque : La somme des degrés d'un graphe est égale au nombre d'arêtes du graphe multiplié par 2.*

On dit qu'un ensemble de sommets forme un **cycle** s'il contient un chemin dont les deux extrémités sont reliées.

```
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
G.add_edge("A","B")
G.add_edge("B","F")
G.add_edge("C","D")
G.add_edge("C","E")
G.add_edge("C","F")
G.add_edge("D","E")
G.add_edge("F","G")
G.add_edge("F","A")
G.add_edge("G","D")

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, arrows = False)

plt.show()
```

Ici, A, B, F forme un cycle de longueur 3. De même, F, C, D, G forme un cycle de longueur 4.

Un **arbre** est un graphe **connexe** qui ne contient pas de **cycle**.

Voici un exemple d'arbre :

```
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_edge("A","B")
G.add_edge("B","C")
G.add_edge("C","D")
G.add_edge("C","E")
G.add_edge("E","F")

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos) # Affichage du nom des sommets
nx.draw_networkx_edges(G, pos, arrows = False)

plt.show()
```

On dit qu'un graphe est **connexe** s'il contient un **arbre couvrant**, c'est à dire s'il est possible d'obtenir un arbre en supprimant certains arrêtes du graphe.

Prenons un graphe en exemple :

```
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
G.add_edge("A","B")
G.add_edge("B","F")
G.add_edge("C","D")
G.add_edge("C","E")
G.add_edge("C","F")
G.add_edge("D","E")
G.add_edge("F","G")
G.add_edge("F","A")
G.add_edge("G","D")

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, arrows = False)

plt.show()
```

Par exemple, en supprimant les arêtes (A,B), (D,G), et (D,E), nous obtenons un arbre :

```
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
# G.add_edge("A","B")
G.add_edge("B","F")
G.add_edge("C","D")
G.add_edge("C","E")
G.add_edge("C","F")
# G.add_edge("D","E")
G.add_edge("F","G")
G.add_edge("F","A")
# G.add_edge("G","D")

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, arrows = False)

plt.show()
```

Nous avons ici un **arbre couvrant** et notre graphe est donc **connexe**.
